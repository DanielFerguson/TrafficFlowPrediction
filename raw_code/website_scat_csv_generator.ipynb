```python
import pandas as pd
from datetime import date,datetime
import numpy as np
import os
import csv 
```

```python
def get_weekday(curr_date):
    curr_date =  datetime.strptime(curr_date,'%d/%m/%Y').date()
    return curr_date.weekday()
```

```python
def generate_data(file):
    df = pd.read_csv("../Data/data/{}".format(file), encoding='utf-8',header=None).fillna(0)
    dates = np.array(df.iloc[:,2])
    new_data = []

    for i in range(0,6):
        week_day_data = []
        for idx,date in enumerate(dates):
            weekday = get_weekday(date)
            if i == weekday:
                week_day_data.append(np.array(df.iloc[idx,3:]))
        week_day_data = np.array(week_day_data).mean(axis=0)
        week_day_data = week_day_data.astype(int)
        new_data.append(week_day_data)
    new_data = new_data[1:] + new_data[:1]
    new_data = np.array(new_data)
    np.savetxt('../Data/website_traffic_data/{}'.format(file), new_data, delimiter=",",fmt='%i')
```

```python
def main():
    files = os.listdir("../Data/data")
    for file in files:
        print(file)
        if not file == "3001_6_data.csv":
            generate_data(file)
        
```

```python
if __name__ == "__main__":
    main()
```
